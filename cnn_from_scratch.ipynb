<img src="static_images/icons/noun-question-mark-869751.png" alt="?" style="width:25px"/> **Question**: Outline the steps you took to get to your final CNN architecture and your reasoning at each step.  

<img src="static_images/icons/noun-answer-3361020.png" alt=">" style="width:25px"/> __Answer:__ 

- I chose a deep CNN with four Convolutional Layers, each followed by Batch Normalization, ReLU activation and Max Pooling.
- This design allows for feature extraction, with each layer capturing complex features. The increase in the number of filters (64 -> 128 -> 256 -> 512) is a common practice to capture more features as the network gets deeper.
- I added Batch Normalization after each convolutional layer and in the classifier. This helps in faster convergence, allows higher learning rates and provides some regularization effect.
- I used Max Pooling after each convolutional block. This reduces spatial dimensions, making the model more computationally efficient and helps in achieving some translation invariance.
- I used two fully-connected layers with dropout in between them. This design allows the model to learn high-level representations for classification. The use of dropout helps prevent overfitting.
- I used ReLU in the convolutional layers and LeakyReLU in the classifier. ReLU is computationally efficient and helps mitigate the vaninshing gradient problem. LeakyReLU in the classifier might help with dead neurons that ReLU can sometimes cause.

Allright, good job getting here! Now it's time to see if all our hard work pays off. In the following cell we will train your model and validate it against the validation set.

Let's start by defining a few hyperparameters. Feel free to experiment with different values and try to optimize your model:

````python
batch_size = 32        # size of the minibatch for stochastic gradient descent (or Adam)
valid_size = 0.2       # fraction of the training data to reserve for validation
num_epochs = 50        # number of epochs for training
num_classes = 50       # number of classes. Do not change this
dropout = 0.5          # dropout for our model
learning_rate = 0.005  # Learning rate for SGD (or Adam)
opt = 'sgd'            # optimizer. 'sgd' or 'adam'
weight_decay = 0.0001     # regularization. Increase this to combat overfitting
````

````python
from src.data import get_data_loaders
from src.train import optimize
from src.optimization import get_optimizer, get_loss
from src.model import MyModel
import torch

# get the data loaders using batch_size and valid_size defined in the previous
# cell
# HINT: do NOT copy/paste the values. Use the variables instead
data_loaders = get_data_loaders(batch_size=batch_size, valid_size=valid_size)

# instance model MyModel with num_classes and drouput defined in the previous
# cell
model = MyModel(num_classes=num_classes, dropout=dropout)
model.load_state_dict(torch.load('./checkpoints/best_val_loss.pt'))
# Get the optimizer using get_optimizer and the model you just created, the learning rate,
# the optimizer and the weight decay specified in the previous cell
optimizer = get_optimizer(model=model, optimizer=opt, learning_rate=learning_rate, weight_decay=weight_decay)

# Get the loss using get_loss
loss = get_loss()

optimize(
    data_loaders,
    model,
    optimizer,
    loss,
    n_epochs=num_epochs,
    save_path="checkpoints/best_val_loss.pt",
    interactive_tracking=True
)
````
